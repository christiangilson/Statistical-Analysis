```r
library(MASS)
```

```r
df <- read.table('oil.txt', col.names = c('spirit','gravity','pressure','distil','endpoint'))
```

## Fit a model with the three best explanatory variables

```r
lm.3 <- lm(spirit ~ gravity + distil + endpoint, data=df)

summary(lm.3)
```

## Manually extracting the fitted and residual values, and then plotting them

```r
lm.3.fitted <- fitted(lm.3)
lm.3.resid <- residuals(lm.3)
```

```r
plot(lm.3.fitted, lm.3.resid)
```

## Or using R's built in plot function with the model as the argument

## Residual plots

1) Plot of standardised residuals Vs fitted values tests the assumption that the random error terms are mean zero and of constant variance

* $e_{i}'$ against $\hat{y}$

* If there is a trend to the data, rather than being randomly dispersed around mean zero, then this indicates that the constant variance assumption does not hold in the model

2) Plot the standardised residuals (which, as the name suggests, is approximately a standard normal distribution) Vs the theoretical quantiles of a standard normal

* This tests the assumption of normality of the random error term in the model (which also applies to the assumption of normality in the dependant variable, $y$.

3) Plot the standardised residuals against the index order that the data was collected to look for any trend that would indicate the data points are not independent.

```r
par(mfrow = c(2,2))
plot(lm.3)
```

## Model adequecy

> Plot the standardised residuals against **EACH EXPLANATORY VARIABLE**

> Expect mean zero, random dispersion

> If there's trend:

* And the variable is already in the model, indicates a higher order variable term is needed (i.e. $x^{2}$)

* If the variable is **NOT** already in the model, then it indicates it should be.

```r
# can easily get the model residuals from:
lm.3.std.res <- stdres(lm.3)
```

## Plotting standardised residuals from the MASS library against **endpoint**

```r
plot(df$endpoint, lm.3.std.res)
```

## Plotting standardised residuals from the MASS library against **distil**

```r
plot(df$distil, lm.3.std.res)
```

## Plotting standardised residuals from the MASS library against **gravity**

```r
plot(df$gravity, lm.3.std.res)
```

## Plotting standardised residuals from the MASS library against **pressure**, which is **NOT** in the model.

> If there was a significant trend here, we'd be tempted to add **pressure** back into the model, despite it not appearing as significant in our goodness-of-fit tests

> Looks random, so we can be happy leaving it out.

```r
plot(df$pressure, lm.3.std.res)
```

# Let's fit a model without the main predictor variable, **endpoint**, then plot the residuals of that model against endpoint to see if we see a trend.

> Note it doesn't pass the null hypothesis test so we wouldn't normally take this any further...

```r
lm.ex.endpoint <- lm(spirit ~ gravity + pressure + distil, data=df)

summary(lm.ex.endpoint)
```

```r
lm.ex.endpoint.std.res <- stdres(lm.ex.endpoint)
```

```r
plot(df$endpoint, lm.ex.endpoint.std.res)
```

## We see clear, clear trend that **endpoint** should be added to the model!
