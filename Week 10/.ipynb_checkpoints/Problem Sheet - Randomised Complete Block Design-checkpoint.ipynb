## Problem Sheet 9

> You can quickly see when something is going to be a blocked design question because the number of replicates per cell is 1.

> Recall idea of blocked design is to essentially run a one-factor experiment, but where you're faced with external variation that you need to control for.

> So you make your blocks as uniform as possible, and run each treatment per block, and set the problem up as a two-way ANOVA, with a treatment factor and a blocking factor

## Q1) Clothing manufacturer

* test the effect of four different dyes
* dependent variable is tensile  strength of denim
* will block by 5 different brands of denim (A->E)

### Practical Implementation

* Dye is our fixed treatment factor
* Brand is our blocking factor
* The order that each of the dyes are used **within each block** should be random to perform a randomised complete block experiment
* We are **assuming that the dye is independent of the brand**, and hense we fit a model similar to two-way ANOVA without the interaction term
* There should only be one observation per treatment per block
* Goal is to try and make the conditions within  each block as uniform  as possible.
* The number of treatment levels of factor A (**dye**):  $a=4$
* The number of blocking levels of factor B (**brand**): $b=5$
* Total number of observations: $ab=20$

### Linear Statistical Model

> $y_{ij} = \mu + \tau_{i} + \beta_{j} + \epsilon_{ij}$,

where $i$ represents the $i$th level of the treatment factor: **dye**, and $j$ represents the $j$th block: **denim**.

$i = 1,2,3,4$, $j=1,2,3,4,5$ (representing A,B,C,D,E).

#### Parameters:

* $\mu$ represents the overall mean
* $\tau_{i}$ represents the treatment effect of the $i$th level of treatment factor A, dye.
* $\beta_{j}$ represents the treatment effect of the $j$th level of the blocking factor B, denim.

### Hypothesis

As dye is a fixed factor, our sum to zero constraint is:

> $\sum^{4}_{i=1}\tau_{i} = 0$

And our null hypothesis is:

> $H_{0}: \tau_{1} = \tau_{2} = \tau_{3} = \tau_{4} = 0$,

where our alternative hypothesis $H_{1}$ is that at least one of the levels in factor A (i.e. one of the dyes) shows a non-zero treatement effect, and that not all of the treatment means are the same.

### F-test

We can reject the $H_{0}$ at the $100\alpha$% level of significance if $F_{A} > F_{a-1,(a-1)(b-1), \alpha} > F_{3,12, \alpha}$



```r
tensile.strength <- c(60,60,62,60,
                      55,54,55,58,
                      61,62,65,62,
                      58,59,60,62,
                      54,57,55,56)
```

```r
brand <- factor(c(rep('A',4),rep('B',4)
                 ,rep('C',4),rep('D',4)
                 ,rep('E',4)))
```

```r
dye <- factor(c(rep(1:4,5)))
```

```r
denim <- data.frame(tensile.strength, dye, brand)
```

```r
rm(tensile.strength,brand,dye)
attach(denim)
```

```r
denim.aov <- aov(tensile.strength ~ dye + brand)

summary(denim.aov)
```

```r
pf(2.376,3,12,lower.tail=F)
```

### Summary of ANOVA

We cannot reject the $H_{0}$ hypothesis that all treatment effects are equal to zero, as the $p$-value associated with the F-test is above a 5% level of significance. The $p$-value is 0.121, meaning that there's a probability of 0.121 that the if we reject the null, that it will actually be true.

The $p$-value associated with the blocking factor, brand, shows that there's significant variation between denim brands. This highlights that if we did not use a blocked design, that the p-value for dye would be much higher, and even less significant.

## Q2) Three tests / treatments for BSE in cattle

* Treatment factor A: tests (T1, T2, T3)
    * $a=3$
* Blocking factor B: cows (1,2,3,4)
    * $b=4$
* Total number of observations $ab=12$

```r
scores <- c(20,23,12,
            29,31,11,
            25,24,8,
            46,51,29)
```

```r
cow <- factor(c(rep(1:4, rep(3,4))))
```

```r
test <- factor(rep(c('T1','T2','T3'),4))
```

```r
bse <- data.frame(scores, test, cow)
```

```r
rm(scores, cow, test)
attach(bse)
```

## Why did experimenter use cows as blocking factor

* Tests may fluctuate cow to cow, so blocking by cow provides a uniform testing environment for each of the treatments, helping to remove external variation.
* The assumption that the cow selected will not "interact" with the treatment seems sound.
* By setting the cows as blocking factors, to ensure a randomised complete block design, the experimenter should randomly assign the **order** of the tests to each cow.

```r
bse.aov <- aov(scores ~ test + cow)

summary(bse.aov)
```

## Summary of ANOVA

We reject the null  hypothesis at the 0.1% level of significance, and reject the alternative hypothesis that at least one of the treatment effects from the three tests is non-zero, and thus that the treatment means are **not** all the same.

```r
plot.design(scores ~ test + cow)
```

```r
model.tables(bse.aov)
```
