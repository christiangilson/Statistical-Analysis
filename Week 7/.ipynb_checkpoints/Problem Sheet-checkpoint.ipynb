# **Q1)** Test scores

```r
scores <- c(33,44,39,38,29,41,39,30,42,44,
            31,33,40,34,31,41,34,28,25,33,
            32,29,34,41,27,26,43,25,35,26,
            39,42,46,42,42,46,39,43,41,38)
```

```r
method <- factor(rep(c('A','B','C','D'), rep(10,4)))
```

```r
teaching <- data.frame(scores, method)
```

```r
rm(scores, method)
attach(teaching)
```

## 1) Let's first take a look at the treatment group and overall means

```r
plot(method, scores)
```

## This next step is basically some `GROUP BY COUNT(*)` style initial analysis

```r
by(scores, method, mean)
```

```r
by(scores, method, var)
```

# Comments:

* Treatment group D has the highest average (both mean and median), and also the lowest variance, both good traits.
    * The lower quartile of D is actually roughly as high as the median of the second best method (A).

* Training group C has the lowest average (both median and mean), and the highest variance, both poor traits.


```r
teaching.aov <- aov(scores ~ method, data=teaching)

summary(teaching.aov)
```

```r
pf(8.343, 3, 36, lower.tail=FALSE)
```

> **Sufficient evidence to reject the null hypothesis (that all treatment groups had the same treatment mean) at the 0.1% significance level**

> Can now move on to pairwise comparisons.

```r
model.tables(teaching.aov, type='means')
```

##  Pairwise t-tests

```r
pairwise.t.test(scores, method, p.adjust.method = 'none')
```

**Suggests that the following treatment effect differences are significantly different:**

* A-B (at the 5% level of significance)
* A-C (at the 5% level of significance)
* B-D (at the 0.01% level of significance)
* C-D (at the 0.01% level of significance)

## Pairwise Tukey tests

> More stringent

> Less likely to reject null when H0 true

> Less likely to reject null when H0 false

```r
TukeyHSD(teaching.aov, conf.level = 0.95)
```

**Interestingly, Tukey's more stringent tests only show three (rather than four with the pairwise t-tests) comparisons between treatment groups to be significantly different, where the difference in treatment means, and thus treatment effects, are non-zero:**

* D-B at the 1% level of significance
* D-C at the 0.1% level of significance
* C-A at the 5% level of significance (only just!)

> The A-B treatment group comparison is no longer seen as significant with the Tukey test.

```r
plot(TukeyHSD(teaching.aov, conf.level = 0.95))
```

# **Q2)** Tyre wear

```r
wear <- c(20.935, 17.123, 29.590, 19.013, 15.919, 28.092, 20.332, 15.285, 28.304,
          18.279, 14.815, 19.973, 21.200, 11.280, 20.096, 19.389, 12.153, 20.477,
          28.535, 37.227, 30.529, 27.998, 38.853, 29.177, 30.073, 40.017, 30.795,
          20.182, 34.340, 29.023, 18.792, 34.707, 28.176, 19.203, 36.307, 28.701)
```

```r
position <- factor(rep(1:4, rep(9,4)))
```

```r
tyres <- data.frame(wear, position)
```

```r
rm(wear, position)
attach(tyres)
```

```r
tyres[c(1:2,10:11,19:20),]
```

```r
by(wear, position, mean)
```

```r
by(wear, position, var)
```

```r
plot(position, wear)
```

## Initial ANOVA

```r
tyres.aov <- aov(wear ~ position, data=tyres)

summary(tyres.aov)
```

**F-test, where $F_{3,32} = 13.74$ has p-value:**

```r
pf(13.74,3,32, lower.tail=FALSE)
```

**Have sufficient evidence from the F-test to reject the null hypothesis at the 0.1% level of significance**. This null hypothesis states that the treatment means between the four treatment groups - i.e. the tyre wear for each position - is the same, and therefore that the treatment effects for each treatment group is zero. The alternative hypothesis, that we accept as a result of rejecting the null, is that at least one of the treatment effects is significantly non-zero.

**Now move on to pairwise comparisions.**

```r
model.tables(tyres.aov, type='means')
```

## Pairwise t-tests

```r
pairwise.t.test(wear, position, p.adjust.method = 'none')
```

**Look to have significant differences between the following treatment group pairs**:

* 1-3  at the 0.1% level of significance
* 2-3 at the 0.1% level of significance
* 1-4 at the 5% level of significance
* 2-4 at the 0.1% level of significance

So we have 4 of the 6 pairwise comparisons showing significant differences.

## Pairwise Tukey tests

```r
tukey.pairwise <- TukeyHSD(tyres.aov, conf.level = 0.95)
tukey.pairwise
```

```r
plot(tukey.pairwise)
```

**Significant differences in the Tukey comparisons**

* 3-1, 3-2, 4-2 all significant at the 5% significance level

* 4-1, which was significant at the 5% level of significance in a pairwise t-test is no longer significant in the Tukey test.

```r
by(wear, position, mean)
```

## Real world interpretation

> Significantly more wear in position 4 than 2

> Significantly more wear in position 3 than 2

> Significantly more wear in position  3 than 1

* 1 = front offside
* 2 =  front nearside
* 3 = rear offside
* 4 = rear nearside

Suggests that both rear tyres wear more quickly than the front near side, and that the rear offside wears more quickly than the front offside.
